# Latihan

```python
# soal 1 = x**(3/4)*4**y
# soal 2 = 3**3/x**y
# soal 3 = 10**(x-4)
```

```python
x = 5
y = 5.1

j1 = x**(3/4)*4**y
j2 = 3**3/x**y
j3 = 10**(x-4)
```

```python
print(j1)
print(j2)
print(j3)
```

```python
# solusi pertama
display(Math('x^{3/4} \\times 4^y = %g' %j1))
```

```python
# solusi kedua
display(Math('\\frac{3^3}{x^y} = %g' %j2))
```

```python
# solusi ketiga
display(Math('10^{x-4} = %g' %j3))
```
