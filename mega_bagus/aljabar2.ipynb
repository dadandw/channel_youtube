```python
import sympy as sym
from IPython.display import Math
```

```python
x,y,z = sym.symbols('x y z')
```

# exponensial

```python
p1 = x**y * x**z
```

```python
display(p1)
display(sym.simplify(p1))
```

```python
p2 = x**y / x**z
display(p2)
display(sym.simplify(p2))
```

```python
p3 = x**y * y**z
display(p3)
display(sym.simplify(p3))
```

# persamaan aljabar

```python
p1 = x*(x+1)
p2 = x
```

```python
display(p1*p2)
```

```python
display(sym.expand(p1*p2))
```

```python
p3 = y**12 + x**8
```

```python
display(p1*p2*p3)
```

```python
display(sym.expand(p1*p2*p3))
```

```python
p4 = sym.sqrt(y)
```

```python
display(p4)
```

```python
display(sym.expand((p1*p2*p3)/p4))
```
