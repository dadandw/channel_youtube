```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
sns.set()
```

```python
# menentukan rataan 2 kelompok
# grup 1 tanpa pupuk
# gru; 2 dengan pupuk
rataan1 = 50
rataan2 = 65

# menentukan standar deviasi (variasi) --sama untuk 2 kelompok
sd = 5

# ukuran sample 2 kelompok
sampel1 = 50
sampel2 = 40
```

```python
# generate bilangan random
test = np.random.randn(sampel1)
test
```

```python
plt.plot(test)
```

```python
sns.distplot(test)
```

```python
# generate data
data1 = np.random.randn(sampel1)*sd + rataan1
data2 = np.random.randn(sampel2)*sd + rataan2

# menggabungkan sample size (jumlah data)
kelompok_data = [sampel1, sampel2]

batas = [np.min(np.hstack((data1, data2))), np.max(np.hstack((data1, data2)))]
```

```python
batas
```

```python
# plot distribusi kedua kelompok (statistik deskriptif)

plt.figure(figsize=(7,5))

a = sns.distplot(data1, hist=False, label='Data1')
a = sns.distplot(data2, hist=False, label='Data2')

a.set(xlabel='Data', ylabel='Density')
plt.legend();
```

```python
# melakukan uji statistik (statistik inferensial)
fig, ax = plt.subplots(1,2, figsize=(8,5), dpi=200)

ax[0].violinplot(data1)
ax[0].plot(1+np.random.randn(sampel1)/10, data1, 'o', color='green')
ax[0].set_ylim(batas)

ax[1].violinplot(data2)
ax[1].plot(1+np.random.randn(sampel2)/10, data2, 'o', color='blue')
ax[1].set_ylim(batas)

# uji t-test
t,p = stats.ttest_ind(data1, data2)

# mencetak hasil t-test
sigtxt = ('', 'TIDAK')
plt.title('Dua kelompok{} berbeda secara signifikan! t({})={}, p={}'.format(sigtxt[int(p>.05)],
                                                                           sum(kelompok_data)-2,
                                                                           np.round(t,2),
                                                                           np.round(p,6)))
plt.show()
```
