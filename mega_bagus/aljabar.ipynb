# Memecahkan persoalan aljabar

```python
x = sym.symbols('x')
```

```python
p = 5*x + 5 - 20
```

```python
sym.solve(p)
```

```python
display(Math('\\text{Solusi dari persamaan }%s \\text{ adalah x = }%g' %(sym.latex(p), sym.solve(p)[0])))
```

```python
p2 = x**2 - 4
sym.solve(p2)
```

```python
jawab = sym.solve(p2)

for i in range(0,len(jawab)):
    print('Solusi #' + str(i+1) + ' adalah ' + str(jawab[i]))
```

```python
y = sym.symbols('y')
```

```python
p3 = x/10 - x * y + 50
```

```python
sym.solve(p3,y)
```
